# Knowledge Graph Construction Demo
This notebook demonstrates:
1. Graph Schema Design
2. Entity-Relationship Modeling
3. Graph Construction in Python using NetworkX


## 1. Graph Schema Design
**Entities:** Person, Company, Project
**Relationships:** WORKS_AT, ASSIGNED_TO, COLLABORATES_WITH


```python
schema = {
    "Person": ["name", "role"],
    "Company": ["name", "domain"],
    "Project": ["title", "status"]
}
schema
```

## 2. Graph Construction Using NetworkX

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

# Add nodes
G.add_node("Alice", type="Person")
G.add_node("Bob", type="Person")
G.add_node("OpenAI", type="Company")
G.add_node("KG Builder", type="Project")

# Add edges (relationships)
G.add_edge("Alice", "OpenAI", relation="WORKS_AT")
G.add_edge("Bob", "OpenAI", relation="WORKS_AT")
G.add_edge("Alice", "KG Builder", relation="ASSIGNED_TO")
G.add_edge("Bob", "KG Builder", relation="COLLABORATES_WITH")

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000)
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
